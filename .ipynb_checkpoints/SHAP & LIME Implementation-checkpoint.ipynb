### Data Pre-processing [1]

The data is a time-series sensor data obtained from NASA vibrational sensors. The data has 4 bearing sensor data as **features - [Bearing 1, Bearing 2, Bearing 3, Bearing 4]** for a time frame as **982 data instances** in total.

```python
dir = "/data/bearing_data1"
dataset = pd.DataFrame()

for filename in os.listdir(dir):
    data = pd.read_csv(os.path.join(dir, filename), sep='\t')
    dataset_mean_abs = np.array(data.abs().mean())
    dataset_mean_abs = pd.DataFrame(dataset_mean_abs.reshape(1,4))
    dataset_mean_abs.index = [filename]
    dataset = dataset.append(dataset_mean_abs)
    
dataset.columns = ['Bearing 1', 'Bearing 2', 'Bearing 3', 'Bearing 4']

dataset.index = pd.to_datetime(dataset.index, format='%Y.%m.%d.%H.%M.%S')
dataset = dataset.sort_index()
dataset.to_csv('Bearing_Dataset.csv')
print("Dataset shape:", dataset.shape)
dataset.head(15)
```

```python
#Train-test Split
train_data = dataset['2004-02-12 10:52:39': '2004-02-16 00:32:39']
test_data = dataset['2004-02-16 00:32:39':]
print("Training dataset shape:", train_data.shape)
print("Test dataset shape:", test_data.shape)
```

### Anomaly Detection [1]

The anomaly detection is performed using thresholding approach. <br>
1. First, model predictions(y_hat) are generated.
2. Prediction error (or) Re-construction error is calculated from actual and predicted values (**|Actual - Predicted|**).
3. A **threshold** value is determined by using the distribution of loss function values in order to detect anomalies [1]. (i.e.) threshold = 0.325.
4. Finally, the prediction error values are compared with the threshold and data points with prediction error greater than threshold are classified as **anomalies** and the others as **normal** data points.

```python
predictions = vanilla_lstm_ts1_model.predict(X_train)
predictions = pd.DataFrame(predictions, columns=train_data.columns)
predictions.index = train_data.index
loss_df = pd.DataFrame(index=train_data.index)
Xtrain = X_train.reshape(X_train.shape[0], X_train.shape[2])
loss_df['Loss'] = np.mean(np.abs(predictions-Xtrain), axis = 1)
plt.figure(figsize=(12,9))
plt.xlim([0.0,.5])
plt.title('Distribution of loss')
plt.grid()
sea.distplot(loss_df['Loss'], bins = 25, kde= True, color = 'Green')
```

```python
def detect_anomaly(X,X_df):
    predictions = vanilla_lstm_ts1_model.predict(X)
    prediction_df = pd.DataFrame(predictions, columns=X_df.columns)
    prediction_df.index = X_df.index
    result_df = pd.DataFrame(index=X_df.index)
    X_test = X.reshape(X.shape[0], X.shape[2])
    result_df['Prediction error'] = np.mean(np.abs(prediction_df-X_test), axis = 1)
    result_df['Threshold'] = 0.325
    result_df['Anomaly'] = result_df['Prediction error'] > result_df['Threshold']
    return result_df

result_train = detect_anomaly(X_train, train_data)
result_test = detect_anomaly(X_test, test_data)
anomaly_results = pd.concat([result_train, result_test])
```

#### Creating data sequences with timestep = 16 for LSTM

```python
#Processing data into sequences of 16 timesteps
def preprocess_data(X_d, y_d, time_steps=1):
    X = []
    y = []
    for i in range(len(X_d) - time_steps):
        a = X_d[i:(i + time_steps)]
        X.append(a)
        b = y_d[i:(i - time_steps)]
        y.append(b)

    return np.array(X), np.array(y)

TIME_STEPS = 16
X_train_ts16, y_train_ts16 = preprocess_data(X_train_transformed, y_train_transformed, TIME_STEPS)
X_test_ts16, y_test_ts16 = preprocess_data(X_test_transformed, X_test_transformed, TIME_STEPS)

print(X_test_ts16.shape, y_test_ts16.shape)
```

### Encoder - Decoder based LSTM Models [5]

1. Here, the vanilla LSTM models are reconfigured into **encoder-decoder** architecture. 
2. It consists of encoder model which reads and encodes the given input sequence into feature vector(repeat vector) and decoder model that accepts feature vector from encoder and try to reconstruct the input sequence.
3. Encoder - Decoder based LSTM Models will predict three-dimensional vector or matrix instead of two-dimensional output vector.

#### LSTM Model with timestep = 1

LSTM model with encoder and decoder using 1 timestep.

### Data loading and pre-processing [7]

1. The data is pre-processed with respect to its time steps as batches. As LSTM accepts only 3D data as input, the actual data is transformed into three dimensional array as $(2620, 250, 25)$. <br>
3. During pre-processing, in order to restructure the data to represent like a supervised learning problem, previous time steps are used as input variables and the next time step is used as output variable which is called **sliding window technique for time-series** <br>
4. Here, the number of previous timesteps is **250** as per config.yaml. Thus our data will be trained and model makes the predictions by processing each 250 timesteps data with 25 features, **ten times**. 
5. Thus the number of predictions will be 10. The output shape will be **$(2620, 10)$ = > (data_samples, n_predictions).**

<br>
The same is demonstrated in the below code. 

```python
train_arr = np.load('D:/DLR/telemanom/data/train/A-1.npy')
data_train = pd.DataFrame(train_arr)
print(data_train.shape)

test_arr = np.load('D:/DLR/telemanom/data/test/A-1.npy')
test = pd.DataFrame(test_arr)
print(test.shape)

#For train data
train=False
data = []
for i in range(len(data_train) - 250 - 10):
    data.append(data_train[i:i + 250 + 10])
data = np.array(data)

assert len(data.shape) == 3

if not train:
    X_train = data[:, :-10, :]
    y_train = data[:, -10:, 0]

print("X train:",X_train.shape)
print("Y train:",y_train.shape)

#For test data
train=False
data = []
for i in range(len(test_arr) - 250 - 10):
    data.append(test_arr[i:i + 250 + 10])
data = np.array(data)

assert len(data.shape) == 3

if not train:
    X_test = data[:, :-10, :]
    y_test = data[:, -10:, 0]
    
print("X test:",X_test.shape)
print("Y test:",y_test.shape)
```

### Model Prediction [7]

1. As explained in the above cell, the model makes predictions with our pre-processed test data.
2. Before predicting the value, actual data is converted into **batches** and each batch is given into model seperately where the model uses previous batch information for current batch processsing and then predicts the output value. <br>
3. Thus each batch contains **260 data samples, 250 timesteps and 25 features** => **(260, 250, 25)**.
4. Our model outputs 10 predictions for 260 data samples as => **(260,10) predictions (y_hat)** for each batch given.

```python
num_batches = int((y_test.shape[0] - 250)/ 70)
for i in range(0, num_batches + 1):
    prior_idx = i * 70
    idx = (i + 1) * 70
    
    if i + 1 == num_batches + 1:
        idx = y_test.shape[0]
    
    X_test_batch = X_test[prior_idx:idx]
    y_test_batch = y_test[prior_idx:idx]
    X_train_batch = X_train[prior_idx:idx]
    y_train_batch = y_train[prior_idx:idx]

y_hat_batch = nasa_model.predict(X_test)
```

### Issues when using SHAP Library

The major issues when using SHAP library are listed below.

**1)**. $Shap.DeepExplainer()$ will only work with models outputting a 2D vector. It also works with LSTM encoder decoder models by flattening the output but it doesn't yield any valid results. If a model with 3D output shape is fed into SHAP library, the below error is thrown.

```python
shap_explainer_enc_dec_lstm_ts16 = shap.DeepExplainer(enc_dec_lstm_ts16_model, X_train_ts16)
shap_values_enc_dec_lstm_ts16 = shap_explainer_enc_dec_lstm_ts16.shap_values(X_test_ts16)
shap.initjs()
shap.force_plot(shap_explainer_enc_dec_lstm_ts16.expected_value[0], shap_values_enc_dec_lstm_ts16[0][0], features)
```

**2)**. $Shap.KernelExplainer()$ doesn't seem to work with 3D input data.

```python
shap_explainer_enc_dec_lstm_ts16_kernel = shap.KernelExplainer(enc_dec_lstm_ts16_model.predict, X_test_ts16)
shap_values_enc_dec_lstm_ts16_kernel = shap_explainer_enc_dec_lstm_ts16_kernel.shap_values(X_test_ts16[451,:,:])
```

**3a)**. SHAP library expects the input feature and output features to be equal. If not, it works but does not yield valid results. A stacked LSTM similar to the previous one is developed with output layer of shape $(None,1)$. When SHAP is applied to this model, it generated results of shape $(1,452,16,4)$.  SHAP library calculates the importance scores for only one feature as the model output layer is of shape $(None,1)$. The value of only one feature (bearing 4) is plotted instead of four features with respect to the model output which will result in incomplete explanations.

```python
shap_explainer_stacked_lstm_ts16_model_2 = shap.DeepExplainer(stacked_lstm_ts16_model_2, X_train_ts16)
shap_values_stacked_lstm_ts16_model_2 = shap_explainer_stacked_lstm_ts16.shap_values(X_test_ts16)

#SHAP visualization
shap.initjs()
shap.force_plot(shap_explainer_stacked_lstm_ts16_model_2.expected_value[0], shap_values_stacked_lstm_ts16_model_2[0][0], features)
```

**3b)**. The above issue is also backed with results from the pre-trained NASA model.

1. The input layer is fed with 25 features, the output layer predicts in shape $(None,10)$ which SHAP considers it as 10 features and not lag observations. But this is not an actual issue with the model as it uses 25 features to make predictions and the output 10 predictions are just lag features and not actual features which creates a conflict when using SHAP library.  $[(260,250,25) \neq (260,10)]$ for SHAP API. <br>

2. Thus the shap values for this model are generated in the shape $(10,1,250,25)$.

```python
#Generating SHAP explanations for only one instance as it is time consuming.
start_time_nasa_model = time.time()
shap_explainer_nasa_model = shap.DeepExplainer(nasa_model, X_train[0:2])
shap_values_nasa_model = shap_explainer_nasa_model.shap_values(X_test_batch[0:1])
end_time_nasa_model = time.time()
time_nasa_model = end_time_nasa_model - start_time_nasa_model
```

### Wrap-up
#### Analysis and Observations

1. LIME doesn't fit for any regression based recurrent models. It only fits for any classifier.
2. SHAP library works for any LSTM model with 2D output layer with multiple steps. It doesn't seem to provide proper or valid results and explanations for encoder-decoder based LSTM models.
3. SHAP library is only supported in TensorFlow 1.X versions, not any other versions.
4. Hyper-parameter tuning is not performed for the above models, so it is unclear that whether tuning the models might change or improve the explanations.
